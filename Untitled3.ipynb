```python
#check the goodness of fit for the following data mapping number of hours a student studies v/s the expected number of hours at 5% los.
#observed data(in hours):8 6 10 7 8 11 9
#Expected data(in hours):9 8 11 8 10 7 6


import numpy as np
from scipy.stats import chi2

# Observed and Expected data
observed = np.array([8, 6, 10, 7, 8, 11, 9])
expected = np.array([9, 8, 11, 8, 10, 7, 6])

# Chi-square statistic
chi_square_stat = np.sum((observed - expected)**2 / expected)

# Degrees of freedom
df = len(observed) - 1

# Critical value at 5% level of significance
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, df)

# Print results
print("Chi-Square Statistic:", chi_square_stat)
print("Degrees of Freedom:", df)
print("Critical Value (5% LOS):", critical_value)

if chi_square_stat < critical_value:
    print("✅ Fit is good: Fail to reject H0")
else:
    print("❌ Fit is not good: Reject H0")


```

```python
#what is the probability of 2 heads in 6 coin tosses? fit the binomial distribution for the coin tosses and plot it


import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parameters
n = 6   # number of tosses
p = 0.5 # probability of heads

# Probability of exactly 2 heads
prob_2_heads = binom.pmf(2, n, p)
print(f"Probability of 2 heads in 6 tosses: {prob_2_heads:.4f}")

# Fit binomial distribution
x = np.arange(0, n+1)
pmf_values = binom.pmf(x, n, p)

# Plot
plt.bar(x, pmf_values, color='skyblue', edgecolor='black')
plt.title("Binomial Distribution: 6 Coin Tosses (p=0.5)")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.xticks(x)
plt.show()
```
